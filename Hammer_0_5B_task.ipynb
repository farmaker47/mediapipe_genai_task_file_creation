```python
TFLITE_MODEL = "/content/ai-edge-torch/hammer_ekv1280.tflite"
TOKENIZER_MODEL = "/content/tokenizer.model"

from mediapipe.tasks.python.genai import bundler

config = bundler.BundleConfig(
    tflite_model=TFLITE_MODEL,
    tokenizer_model=TOKENIZER_MODEL,
    start_token="<|im_start|>user\n",     # begin user turn
    stop_tokens=["<|im_end|>"],            # end of any turn
    output_filename="/content/hammer2p1_05b.task",
    prompt_prefix="",                      # prefix is handled via start_token
    prompt_suffix="<|im_end|>\n<|im_start|>assistant\n",  # close user, open assistant
)
bundler.create_bundle(config)
```
