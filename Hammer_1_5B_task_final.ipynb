```python
# --- Hammer 2.1 Model Configuration ---
# Based on https://huggingface.co/MadeAgents/Hammer2.1-1.5b/blob/main/tokenizer_config.json

# 1. Path to your converted Hammer 2.1 TFLite model file.
TFLITE_MODEL = "/content/ai-edge-torch/hammer_q8_ekv1280.tflite"

# 2. Path to the tokenizer model file for Hammer 2.1.
TOKENIZER_MODEL = "/content/tokenizer.model"

# 3. Define the special tokens based on Hammer 2.1's configuration.
START_TOKEN = "<s>"
STOP_TOKENS = ["<|im_end|>", "</s>"]

# 4. Define the prompt format for the bundler.
#    This tells the model how the conversation is structured.
PROMPT_PREFIX = "<|im_start|>user\n"
PROMPT_SUFFIX = "<|im_end|>\n<|im_start|>assistant\n"

# 5. Define the output filename for the bundle.
OUTPUT_FILENAME = "/content/hammer2_1.task"

# --- Bundler Script ---

from mediapipe.tasks.python.genai import bundler

print(f"Creating bundle for model: {TFLITE_MODEL}")

config = bundler.BundleConfig(
    tflite_model=TFLITE_MODEL,
    tokenizer_model=TOKENIZER_MODEL,
    start_token=START_TOKEN,
    stop_tokens=STOP_TOKENS,
    prompt_prefix=PROMPT_PREFIX,
    prompt_suffix=PROMPT_SUFFIX,
    output_filename=OUTPUT_FILENAME,
)

bundler.create_bundle(config)

print(f"Successfully created bundle at: {OUTPUT_FILENAME}")
```
